```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
infected_fraction = pd.read_csv("infected_fraction_hist.csv", names=[f"group {i+1}" for i in range(40)])
infected_fraction.tail()
```

```python
recovered_fraction = pd.read_csv("recovered_fraction_hist.csv", names=[f"group {i+1}" for i in range(40)])
recovered_fraction.tail()
```

```python
population = pd.read_csv("population_hist.csv", names=[f"group {i+1}" for i in range(40)])
population.tail()
```

```python
infected_fraction.plot(figsize = (20, 15))
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=14)
plt.show()
```

```python
recovered_fraction.plot(figsize = (20, 15))
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=14)
plt.show()
```

```python
population.plot(figsize = (20, 15))
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=14)
plt.show()
```
